## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df=attrition_df[['Attrition','Department']]
y_df
```

```python
attrition_df.columns
```

```python
# Create a list of at least 10 column names to use as X data
X_columns=['Age', 'DistanceFromHome',
       'Education', 'EducationField', 'EnvironmentSatisfaction', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobSatisfaction',
       'MaritalStatus', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
       'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']


# Create X_df using your selected columns
X_df=attrition_df[X_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
X_df['MaritalStatus'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['Divorced', 'Single', 'Married']])
X_df['MaritalStatusEncoded']=encoder.fit_transform(X_df[['MaritalStatus']])
X_df=X_df.drop(columns='MaritalStatus')
X_df['MaritalStatusEncoded']
```

```python
X_df['EducationField'].unique()
```

```python
education_categories=['Other', 'Life Sciences', 'Medical', 'Marketing', 'Technical Degree', 'Human Resources']
encoder = OrdinalEncoder(categories=[education_categories])
X_df['EducationFieldEncoded']=encoder.fit_transform(X_df[['EducationField']])
X_df=X_df.drop(columns='EducationField')
X_df['EducationFieldEncoded']
```

```python
X_df['OverTime'].unique()
```

```python
encoder=OrdinalEncoder(categories=[['No','Yes']])
X_df['OvertimeEncoded']=encoder.fit_transform(X_df[['OverTime']])
X_df=X_df.drop(columns='OverTime')
X_df['OvertimeEncoded']
```

```python
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
```

```python
# Create a StandardScaler
scaler=StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled_df
```

```python
y_test['Department'].unique()
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
dep_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit the encoder to the training data
dep_encoder.fit(y_train['Department'].values.reshape(-1, 1))

# Transform the 'Department' column in both training and testing data
y_train_dep_encoded = dep_encoder.transform(y_train['Department'].values.reshape(-1, 1))
y_test_dep_encoded = dep_encoder.transform(y_test['Department'].values.reshape(-1, 1))
```

```python
# Create a OneHotEncoder for the Attrition column
attr_encoder=OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
# Fit the encoder to the training data
attr_encoder.fit(y_train['Attrition'].values.reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attr_encoded=attr_encoder.transform(y_train['Attrition'].values.reshape(-1,1))
y_test_attr_encoded=attr_encoder.transform(y_test['Attrition'].values.reshape(-1,1))
```

```python
y_train_dep_df=pd.DataFrame(y_train_dep_encoded, columns = dep_encoder.get_feature_names_out())
y_test_dep_df=pd.DataFrame(y_test_dep_encoded, columns = dep_encoder.get_feature_names_out())
y_train_attr_df=pd.DataFrame(y_train_attr_encoded, columns = attr_encoder.get_feature_names_out())
y_test_attr_df=pd.DataFrame(y_test_attr_encoded, columns = attr_encoder.get_feature_names_out())
y_train_encoded=pd.concat([y_train_dep_df,y_train_attr_df],axis=1)
y_test_encoded=pd.concat([y_test_dep_df,y_test_attr_df],axis=1)
y_train_encoded
```

```python
y_train_encoded['HR']=y_train_encoded['x0_Human Resources']
y_train_encoded['RaD']=y_train_encoded['x0_Research & Development']
y_train_encoded['Sales']=y_train_encoded['x0_Sales']
y_train_encoded['Attrition']=y_train_encoded['x0_Yes']
y_train_encoded=y_train_encoded.drop(columns=['x0_Human Resources','x0_Research & Development','x0_Sales','x0_Yes'])
y_train_encoded
```

```python
y_test_encoded['HR']=y_test_encoded['x0_Human Resources']
y_test_encoded['RaD']=y_test_encoded['x0_Research & Development']
y_test_encoded['Sales']=y_test_encoded['x0_Sales']
y_test_encoded['Attrition']=y_test_encoded['x0_Yes']
y_test_encoded=y_test_encoded.drop(columns=['x0_Human Resources','x0_Research & Development','x0_Sales','x0_Yes'])
y_test_encoded
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
shape_x=X_train_scaled_df.shape[1]

# Create the input layer
input_layer=layers.Input(shape=shape_x,name='input_features')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dep_dense=layers.Dense(16,activation='relu')(shared_layer2)

# Create the output layer
dep_output=layers.Dense(len(y_train_dep_df.columns),activation='sigmoid',name='dep_output')(dep_dense)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_dense=layers.Dense(16,activation='relu')(shared_layer2)

# Create the output layer
attr_output=layers.Dense(len(y_train_attr_df.columns),activation='sigmoid',name='attr_output')(attr_dense)
```

```python
# Create the model
model=Model(inputs=input_layer,outputs=[dep_output,attr_output])

# Compile the model
model.compile(optimizer='adam',
             loss={'dep_output':'binary_crossentropy','attr_output':'binary_crossentropy'},
             metrics={'dep_output':'accuracy','attr_output':'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled_df,
         {'dep_output':y_train_dep_encoded,'attr_output':y_train_attr_encoded},
         epochs=50,
         batch_size=16,
         validation_split=0.2)
```

```python
# Evaluate the model with the testing data
results=model.evaluate(X_test_scaled_df,
         {'dep_output':y_test_dep_encoded,'attr_output':y_test_attr_encoded})
print(results)
```

```python
# Print the accuracy for both department and attrition
print(f"Department predictions accuracy: {results[3]}")
print(f"Attrition predictions accuracy: {results[4]}")
```
